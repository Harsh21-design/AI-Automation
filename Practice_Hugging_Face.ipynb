#using Text-Generation Pipeline

```python
from transformers import pipeline

mytext = pipeline("text-generation", model="gpt2")

mytext("How to use AI")
```

```python
from transformers import pipeline

mytext = pipeline("text-generation", model="gpt2")

mytext(
    "What is chatgpt?",
    max_length = 20,
    num_return_sequences = 2, #number of sentences
    )
```

#using sentiment-analysis pipeline

```python
from transformers import pipeline
mysentiments = pipeline("sentiment-analysis")

mysentiments("I love to play video games")
```

```python
mysentiments = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
    )

mysentiments("Aaj mujhe khana bnana pdega")
```
