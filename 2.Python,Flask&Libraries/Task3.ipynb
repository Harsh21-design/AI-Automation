```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Missing.csv")
```

```python
dataset.info()
```

```python
dataset.isna().any()
```

```python
print(dataset)
```

```python
# Convert dataset into datframe
df = pd.DataFrame(dataset)
```

```python
df["Name"]=df["Name"].fillna("Aarti")
df
```

```python
df["Gender"]=df["Gender"].fillna("Male")
df
```

```python
df["Class"]=df["Class"].fillna(df['Class'].mean())
df["Age"]=round(df["Age"].fillna(df['Age'].mean()),2)
df["Weight"]=round(df["Weight"].fillna(df['Weight'].mean()),2)
df["Height"]=df["Height"].fillna(df['Height'].mean())
df
```

```python
df.info()
```

```python
summaryMissing = df.describe()
print(summaryMissing)
```

```python
summaryMissing.to_csv("summaryMissing.csv")
```
